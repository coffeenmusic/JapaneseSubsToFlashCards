# Create Anki Deck

```python
import genanki

WORD_COUNT = 10
deck_name = sub_file.split('.')[0]
```

```python
model_id = random.randrange(1 << 30, 1 << 31)

my_model = genanki.Model(
  model_id,
  'Simple Model',
  fields=[
    {'name': 'Question'},
    {'name': 'Answer'},
  ],
  templates=[
    {
      'name': 'Card 1',
      'qfmt': '{{Question}}',
      'afmt': '{{FrontSide}}<hr id="answer">{{Answer}}',
    },
  ])

deck_id = random.randrange(1 << 30, 1 << 31)
deck = genanki.Deck(
  deck_id,
  deck_name)
```

```python
import jisho
from urllib.parse import quote  
```

```python
answer = jisho.search(quote('お前'))[0]
```

```python
for word, _ in word_counts.most_common()[:WORD_COUNT]:
    answer = jisho.search(quote(word))[0]
    #answer.ja[0].reading
    #answer.en[0].meaning
    card = genanki.Note(model=my_model, fields=[word, str(answer)])
    deck.add_note(card)
```

```python
genanki.Package(deck).write_to_file(f'{deck_name}.apkg')
```
